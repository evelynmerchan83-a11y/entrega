```python

class Estudiante:
    def __init__(self, nombre, apellido, cedula, email, semestre, materia):
        # Atributos privados
        self.__nombre = nombre
        self.__apellido = apellido
        self.__cedula = None
        self.__email = email
        self.__semestre = None
        self.__materia = materia

        # Usamos los setters para validar los datos iniciales
        self.cedula = cedula
        self.semestre = semestre

    # --- Propiedades para NOMBRE ---
    @property
    def nombre(self): return self.__nombre
    @nombre.setter
    def nombre(self, v): self.__nombre = v

    # --- Propiedades para APELLIDO ---
    @property
    def apellido(self): return self.__apellido
    @apellido.setter
    def apellido(self, v): self.__apellido = v

    # --- Propiedad CEDULA (Validación 10 dígitos) ---
    @property
    def cedula(self):
        return self.__cedula

    @cedula.setter
    def cedula(self, valor):
        if isinstance(valor, str) and len(valor) == 10 and valor.isdigit():
            self.__cedula = valor
        else:
            print(f"Error: La cédula '{valor}' debe tener exactamente 10 dígitos numéricos.")

    # --- Propiedades para EMAIL ---
    @property
    def email(self): return self.__email
    @email.setter
    def email(self, v): self.__email = v

    # --- Propiedad SEMESTRE (Validación 1-10) ---
    @property
    def semestre(self):
        return self.__semestre

    @semestre.setter
    def semestre(self, valor):
        if 1 <= valor <= 10:
            self.__semestre = valor
        else:
            print(f"Error: El semestre '{valor}' no es válido. Debe estar entre 1 y 10.")

    # --- Propiedades para MATERIA ---
    @property
    def materia(self): return self.__materia
    @materia.setter
    def materia(self, v): self.__materia = v

    # --- Método Especial ---
    def __str__(self):
        """Representación basada en el diccionario interno."""
        return str(self.__dict__)

# --- Ejemplo de uso ---

# Intento con datos correctos
estudiante1 = Estudiante("Ana", "Paz", "1723456789", "ana@mail.com", 3, "Programación")
print(estudiante1)

# Intento con datos inválidos (activará las validaciones)
estudiante2 = Estudiante("Luis", "Vega", "123", "luis@mail.com", 15, "Cálculo")
print(estudiante2)
```
